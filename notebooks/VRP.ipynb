```python
origin = 2
points = np.random.randint(0, 100, n)
pt_indexes = itertools.product(*[range(m) for i in range(n)])
combinations = []
for indexes in pt_indexes:
    cups = [[origin] for _ in range(m)]
    for i, index in enumerate(indexes):
        cups[index].append(points[i])
    cups[index].append(origin)
    combinations.append(cups)
```

```python
p = itertools.permutations(range(n*m), n)
```

```python
def map_path(p):
    paths = [[] for _ in range(m)]
    for i, j in enumerate(p):
        paths[j // n].append(i)
    return paths
```

```python
pp = map(map_path, p)
```
