## Gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(w):
    return (w - 10) ** 2  # w^2 -20*w + 10
```

```python
def df(w):
    return 2 * w - 20
```

```python
w = np.random.randint(0, 100)
learning_rate = 0.1
e = 0.0
cost = 0
old_w = np.random.randint(0, 200)
```

```python
while abs(f(old_w) - f(w)) > e:
    gradw = df(w)
    old_w = w
    w -= learning_rate * gradw
    cost = f(old_w) - f(w)
    print('cost: {}, gradient: {}, w: {}'.format(cost, gradw, w))
```

## draw function graph

```python
def draw(func):
    w = np.linspace(-10,30,1000)
    f = func(w)
    plt.plot(w, f)
```

```python
draw(f)
```
